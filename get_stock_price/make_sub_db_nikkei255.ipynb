### データベースのコンストラクト 

```python
db_path = Path("db/stock_db") / Path("stock.db")
stock_db = StockDatabase(db_path)
```

## 利用する銘柄名 

```python
ticker_codes_df = pd.read_csv(Path("portfolio/rl_base/nikkei225.csv"), header=0)  # 自分で作成
ticker_codes = ticker_codes_df["code"].values.astype(str).tolist()
ticker_codes
```

### 新しいサブデータベースの作成 

```python
new_db_path = Path("db/sub_stock_db") / Path("nikkei_255_stock_v2.db")

jst_timezone = timezone("Asia/Tokyo")
start_datetime = jst_timezone.localize(datetime.datetime(2020,11,1,0,0,0))
end_datetime = jst_timezone.localize(datetime.datetime(2021,2,1,0,0,0))
#end_datetime = get_next_workday_jp(start_datetime, days=11)  # 営業日で一週間(5日間)

stock_list = ticker_codes

new_db = stock_db.make_db(new_db_path, 
                          stock_names=stock_list,
                          start_datetime=start_datetime,
                          end_datetime=end_datetime
                         )
```

```python
stock_names = ["4755"]
stock_df = new_db.search_span(stock_names,
                              start_datetime=start_datetime,
                              end_datetime=end_datetime,
                              freq_str="T",
                              to_tokyo="True"
                             )
```
